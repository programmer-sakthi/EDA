```python
import matplotlib.pyplot as plt
```

```python
# create a plot
plt.plot([1,2,3,4], [1,4,9,16])
plt.show()
```

```python
# customize the plot
plt.plot([1,2,3,4], [1,4,9,16], 'ro')  # red circles
plt.title('Square Numbers')
plt.xlabel('Value')
plt.ylabel('Square')
plt.grid(True)
plt.show()
```
