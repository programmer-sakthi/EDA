```python
# Part B question
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# overwrite the previous incorrect read_csv and read the Excel file properly
try:
    df = pd.read_excel('DA314_S7_EmployeePerformance_Data_Concept.xlsx', engine='openpyxl')
except Exception:
    df = pd.read_excel('DA314_S7_EmployeePerformance_Data_Concept.xlsx')

print(f"Loaded {df.shape[0]} rows and {df.shape[1]} columns")
df.head()
```

```python
# Promotion status pie chart 
plt.figure(figsize=(6,6))
promotion_counts = df['is_promoted'].value_counts()
plt.pie(promotion_counts, labels=['Not Promoted', 'Promoted'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Promotion Status Distribution')
plt.show()
```

```python
# gender distribution pie chart
plt.figure(figsize=(6,6))
gender_counts = df['gender'].value_counts()
plt.title("Gender Distribution")
plt.pie(gender_counts,labels=['male','female'],autopct='%1.1f%%',startangle=90, colors=['#ffcc99','#99ff99'])
```

```python
# bar chart for department vs workforce diversity
plt.figure(figsize=(10,6))
dept_diversity = df.groupby('department')['employee_id'].nunique().sort_values(ascending=False)
sns.barplot(y=dept_diversity.index, x=dept_diversity.values , color='red')
plt.title('Department vs Workforce Diversity')
```

```python
# bar chart for region vs workforce diversity
plt.figure(figsize=(10,6))
dept_diversity = df.groupby('region')['employee_id'].nunique().sort_values(ascending=False)
sns.barplot(x=dept_diversity.index, y=dept_diversity.values , color='pink')
plt.xticks(rotation=90)
plt.title('Department vs Workforce Diversity')
```
