```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
orders=pd.read_csv("DA314_S4_GlamNest_Data_Practice.csv")
print(orders.head())
```

```python
plt.hist(orders['final_amount'] , edgecolor = 'black' , bins=10 , color = '#a5daf0')
plt.xlabel('final_amount (INR)')
plt.ylabel('frequency')
plt.show()
```

```python
sns.kdeplot(data = orders , x = 'service_rating' , fill = True  , edgecolor = 'green' , color = 'green')
```

```python
sns.boxplot(data=orders , x='category' , y='final_amount')
plt.xlabel('Category')
plt.ylabel('Final Amount ( INR ) ')
plt.title('Boxplot of Final Amount by Category')
plt.xticks(rotation=45)
plt.show()
```
